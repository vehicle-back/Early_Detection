```python
def calculate_average_with_mask(df, mask_df, number):
    results = []
    
    # 根据mask_df将df中mask为1的值替换为NaN
    df_with_nan = df.mask(mask_df == 1)
    
    for i in tqdm(range(len(df))):
        current_row = df.iloc[i]
        day_of_week = current_row['day_of_week']
        time_of_day = current_row['time_of_day']
        
        # 获取当前行之前的同一星期几和时间段的行
        previous_rows = df_with_nan[(df['day_of_week'] == day_of_week) & 
                                    (df['time_of_day'] == time_of_day) & 
                                    (df.index < df.index[i])]
        
        # 获取当前行之后的同一星期几和时间段的行
        subsequent_rows = df_with_nan[(df['day_of_week'] == day_of_week) & 
                                      (df['time_of_day'] == time_of_day) & 
                                      (df.index > df.index[i])]
        
        # 选择前两行或后两行
        if len(previous_rows) >=  number:
            average_rows = previous_rows.tail( number)
        elif len(subsequent_rows) >=  number and len(previous_rows)< number:
            average_rows = subsequent_rows.head( number)
        else:
            raise ValueError('Wrong! ')
        
        # 如果依然不足两行数据，则结果为NaN
        #if average_rows.dropna(how='all').shape[0] ==0:
        #    avg_values = [np.nan] * len(df.columns[:-2])
        #else:
        avg_values = average_rows.iloc[:, :-2].mean(skipna=True).values
        
        results.append(avg_values)
    
    # 创建一个新的DataFrame来保存结果
    avg_df = pd.DataFrame(results, columns=df.columns[:-2], index=df.index)
    return avg_df
```

```python
mask_avg =  calculate_average_with_mask(spd_record_all, mask, 3)
```

```python
def fill_nan_with_nearest_date(df):
    # Ensure 'date' column exists in the DataFrame
    if 'date' not in df.columns:
        df['date'] = df.index.date
    
    # Group by 'date'
    grouped = df.groupby('date')

    # Apply forward fill and backward fill within each group
    filled_df = grouped.apply(lambda group: group.ffill().bfill())

    # Drop the 'date' column as it is no longer needed
    filled_df = filled_df.drop(columns=['date'])

    filled_df = filled_df.reset_index(level='date', drop=True)
    
    return filled_df
```

```python
no_nan_mask_avg = fill_nan_with_nearest_date(mask_avg)
```
